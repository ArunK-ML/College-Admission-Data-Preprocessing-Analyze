```python
import pandas as pd
```

```python
df = pd.read_csv('/content/College_Admission_data.csv')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df.fillna(100)   # na = fill at the null value
```

```python
df
```

```python
df.fillna(method = "bfill") # backward fill
```

```python
df.fillna(method = "ffill") # ffill - forward fill
```

```python
df.ffill() # In New version
```

```python
df['hsc']=df['hsc'].fillna(method="bfill") # Changing the hsc column alone
```

```python
df
```

```python
df['SSLC'].fillna(df['SSLC'].mean(),inplace=True)
```

```python
df
```

```python
df.dropna() # Dropping the records which has null values
```

```python
df.dropna(axis=1) # Dropping the columns which has null values # Don't try this
```

```python
df.dropna(how= "any") # by default this will erase the records which has null values
```

```python
df.dropna(how= "all") # if you have null values in all the rows
```

```python
df.dropna(how = "any",subset = ["hsc","SSLC","arrears"])
```

```python
df.dropna(how="all",subset = ["hsc","SSLC","arrears"])
```
